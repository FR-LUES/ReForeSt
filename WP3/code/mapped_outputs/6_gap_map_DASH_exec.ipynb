## Split raster to export from DASH


```r
england_gap_map_full <- rast(paste0(path_export_eng_DASH, "england_VOM_gaps_clip.tif"))
```

```r
# create 16 tiles
tile_dims <- ceiling(dim(england_gap_map_full)[1:2] / c(4,4))
```

```r
# make tiles 
makeTiles(england_gap_map_full, tile_dims, buffer = 10, filename = "/tmp/england_VOM_gaps_clip_.tif")
```

```r
# Copy to DBFS

for (i in 1:16) {
  
  file.copy(paste0("/tmp/england_VOM_gaps_clip_", i, ".tif"), paste0(path_export_eng_DASH, "england_VOM_gaps_clip_", i, ".tif"))
  file.remove(paste0("/tmp/england_VOM_gaps_clip_", i, ".tif"))

}
```
